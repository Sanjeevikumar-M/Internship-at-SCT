```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/world_population_by_country_2023.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
# Display column names to understand the dataset
print(df.columns)
```

```python
# Convert 'population' to numeric (handle any errors)
df['population'] = pd.to_numeric(df['population'], errors='coerce')

# Sort by population and take the top 10
top_10_countries = df.nlargest(10, 'population')

# Create bar chart
plt.figure(figsize=(10, 5))
plt.bar(top_10_countries['country'], top_10_countries['population'], color='blue')

# Labels and title
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2023)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'median_age' to numeric
df['median_age'] = pd.to_numeric(df['median_age'], errors='coerce')

# Create histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['median_age'], bins=20, kde=True, color='green')
plt.xlabel("Median Age")
plt.ylabel("Number of Countries")
plt.title("Distribution of Median Age Across Countries")
plt.show()
```

```python
# Get top 10 countries by median age
top_10_median_age = df.nlargest(10, 'median_age')

# Create bar chart
plt.figure(figsize=(10, 5))
plt.bar(top_10_median_age['country'], top_10_median_age['median_age'], color='orange')
plt.xlabel("Country")
plt.ylabel("Median Age")
plt.title("Top 10 Countries with Highest Median Age")
plt.xticks(rotation=45)
plt.show()
```

```python
# Define estimated continent mapping (modify as needed)
continent_mapping = {
    'China': 'Asia', 'India': 'Asia', 'United States': 'North America',
    'Indonesia': 'Asia', 'Pakistan': 'Asia', 'Brazil': 'South America',
    'Nigeria': 'Africa', 'Bangladesh': 'Asia', 'Russia': 'Europe',
    'Mexico': 'North America', 'Japan': 'Asia', 'Ethiopia': 'Africa'
}

# Create a new continent column (default as 'Other')
df['continent'] = df['country'].map(continent_mapping).fillna('Other')

# Group by continent and sum population
continent_population = df.groupby('continent')['population'].sum()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(continent_population, labels=continent_population.index, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple', 'orange', 'gray'])
plt.title("World Population Share by Continent (Estimated)")
plt.show()
```
